```python
import pandas as pd
```

```python
df = pd.read_csv('coronavirus.csv') # carregar o arquivo
df.head(10) # os 10 primeiros
```

```python
df.tail(10) # os 10 ultimos
```

```python
import seaborn as sns # biblioteca usada para zerar gráficos
import matplotlib.pyplot as plt
```

```python
 plt.figure(figsize=(7, 6))
sns.barplot(x=df['regiao'], y=df['obitosNovos'],data=df)
```

```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
```

```python
trace = go.Scatter(x = df['estado'],
                   y = df['obitosNovos'],
                   mode = 'markers')
data = [trace]
layout = go.Layout(title = 'Gráfico : Obitos Por Estados',
                   xaxis = {'title': 'Estados'},
                   yaxis = {'title': 'Obitos'})
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
trace = go.Bar(x = df['regiao'],
               y = df['obitosNovos'])
data = [trace]
layout = go.Layout(title = 'Gráfico : Obitos Por Região',
                   xaxis = {'title': 'Regiao'},
                   yaxis = {'title': 'Obitos'})
fig = go.Figure(data=data, layout=layout)
py.iplot(data)
```

```python
trace2 = go.Bar(x = df['regiao'],
                y = df['casosNovos'])
data = [trace2]
layout = go.Layout(title = 'Gráfico : Obitos Por Região',
                   xaxis = {'title': 'Região'},
                   yaxis = {'title': 'casosNovos'})
fig = go.Figure(data=data, layout=layout)
py.iplot(data)
```
