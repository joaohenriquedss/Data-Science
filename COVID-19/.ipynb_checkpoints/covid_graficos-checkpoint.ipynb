## A Biblioteca plotly √© usada para criar graficos animados

```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
```

```python
trace = go.Scatter(x = df['estado'],
                   y = y,
                   mode = 'markers')
data = [trace]
layout = go.Layout(title = 'Gr√°fico : Obitos Por Estados',
                   xaxis = {'title': 'Estados'},
                   yaxis = {'title': 'Obitos'})
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
trace = go.Bar(x = x,
               y = y)
data = [trace]
layout = go.Layout(title = 'Gr√°fico : Obitos Por Estados',
                   xaxis = {'title': 'Estados'},
                   yaxis = {'title': 'Obitos'})
fig = go.Figure(data=data, layout=layout)
py.iplot(data)
```
